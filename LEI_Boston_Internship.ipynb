```python
import pandas as pd
df = pd.read_csv('Dictionary.csv') # Read the dictionary of possible meaning( product-wise) with phrases in queries.
```

```python
df.head()
```

```python
df_data =pd.read_csv('Document.csv') # Read the queries set according to dates
```

```python
df_data.head()
```

```python
meaning_dict={}
for i,row in df.iterrows():
    meaning_dict[row['Keywords in Queries']]=row['Meaning ']
```

```python
meaning_dict
```

```python
df1=pd.DataFrame(columns=['Date','Inq','Matches']) # Create empty dataframe where we will store our Inquiries with matches
# Phrase matching
for i,row in df_data.iterrows():
#     print(i)
    #print(row['Date'],row['Inq'])
    if row['Inq']:
        all_keys=set()
        #all_keys.add(row['Date'])
        for phrase in meaning_dict:
            if phrase in row['Inq']:
                #all_keys.add(row['Date'])
                all_keys.add(meaning_dict[phrase])
        df1 = df1.append({'Date': row['Date'],'Inq':row['Inq'],'Matches':str(all_keys).replace('{','').replace('}','') }, ignore_index=True)
        #print(p)       
        
```
